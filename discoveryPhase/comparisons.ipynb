```python
def plot_compare_all(co2, cc, facen, title):
    fig, (ax1, ax2)= plt.subplots(1, 2, figsize=(15, 5))

    x =np.linspace(9,9.6,len(co2))

    ax1p = ax1.twinx()
    ax1.plot(x,co2, label='co2')
    ax1.set_ylim(300, 900)
    ax1.set_yticks(np.linspace(300, 900, 6))
    ax1p.plot(x, cc, color="orange", label='crowdcount')
    ax1p.set_ylim(0, 150)
    ax1p.set_yticks(np.linspace(0, 150, 6))
    handles1, labels1 = ax1.get_legend_handles_labels()
    handles1p, labels1p = ax1p.get_legend_handles_labels()
    handles1 = handles1 + handles1p
    labels1 = labels1 + labels1p
    ax1.legend(handles1, labels1)

    ax2p = ax2.twinx()
    ax2.plot(x,co2, label='co2')
    ax2.set_ylim(300, 900)
    ax2.set_yticks(np.linspace(300, 900, 6))
    ax2p.plot(x, facen, color="red", label='facentration')
    ax2p.set_ylim(30, 100)
    ax2p.set_yticks(np.linspace(30, 100, 6))
    handles2, labels2 = ax2.get_legend_handles_labels()
    handles2p, labels2p = ax2p.get_legend_handles_labels()
    handles2 = handles2 + handles2p
    labels2 = labels2 + labels2p
    ax2.legend(handles2, labels2)

    ax1.set_title(title)

plot_compare_all(co2_crypto, cc_crypto, facen_crypto, "Crypto 9:00am - 10:00am")
```

```python
from itertools import cycle
def plot_compare_metrics(metrics, labels, title):
    fig, ax = plt.subplots(figsize=(20, 5))
    colors = cycle(plt.cm.tab10.colors)
    for metric, label in zip(metrics, labels):
        color = next(colors)
        x = np.linspace(1,len(metric),len(metric))
        for y in range(6, 24, 6):
            if y != 6:
                label=None
            ax.plot(x[y-6: y], metric[y-6: y], label=label, linewidth=4, color=color)
            ax.scatter(x[y-6: y], metric[y-6: y], linewidth=0.1, color="black", zorder=10, alpha=0.5)
        ax.set_xticks([1,7,13], [1, 2, 3])
    ax.set_ylabel(title)
    ax.set_xlabel("lecture sequence")
    ax.legend()
```
